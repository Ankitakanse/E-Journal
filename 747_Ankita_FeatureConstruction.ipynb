```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('Data/train.csv')[['Age','Pclass','SibSp','Parch','Survived']]

df.head() # parch parent child
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
x=df.iloc[:,0:4]
y=df.iloc[:,-1]
```

```python
x.head()
```

```python
cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=20) # before validation we have 20 different values so we are getting diff accuracy.
```

```python
# cv = cross validation
np.mean(cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=20))
```

# Applying Feature Construction

```python
x['Family_size']=x['SibSp']+x['Parch']+1 # family + that person who is travelling (1)
```

```python
x.head()
```

```python
def myfunc(num):
    if num==1:
        #alone
        return 0
    elif num>1 and num<=4:
        #small family
        return 1
    else:
        #large family
        return 2
    
```

```python
myfunc(4)
```

```python
x['Family_type']=x['Family_size'].apply(myfunc)
```

```python
x.head()
```

```python
x.drop(columns=['SibSp','Parch','Family_size'],inplace=True)
```

```python
x.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=20))
```

so we can see that previously accuracy was 0.69 and now it has been changed.
