```python
x=df['Temperature'].values.reshape(-1,1) # values 1-d array after that reshape to convert it into 1-d
y=df['Revenue'].values.reshape(-1,1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['Temperature'],df['Revenue']) # scatter plot
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.1)
```

```python
x_train.shape # input data y_train output data
```

```python
x_test.shape
```

```python
x_train # random data
```

```python
from sklearn.linear_model import LinearRegression # for multiple use this only
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train) # model fit finding parameter traininig
```

```python
y_pred=lr.predict(x_test) # evaluate x data testing
```

```python
pred=pd.DataFrame(y_test,columns=['y_test']) # prediction should not be exact 
pred['y_hat']=y_pred # predicted revenue
pred['residuals']=y_test-y_pred # error
```
